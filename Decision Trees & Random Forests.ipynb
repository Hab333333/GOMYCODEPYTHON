```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import graphviz
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier 
from sklearn import metrics 
ts = pd.read_csv('C:/Users/DELL/Downloads/titanic-passengers.csv', sep=";")
ts.info()
ts=ts.dropna(subset=['Age'])
print(ts["Cabin"].value_counts())
print(ts["Embarked"].value_counts())
ts["Cabin"].fillna("C23 C25 C27 ", inplace = True)
ts["Embarked"].fillna("S", inplace = True)
cleanup={"Survived":{"Yes":1, "No": 0}} 
ts.replace(cleanup, inplace=True)
cleanup_1={"Sex":{"male":1, "female": 0}} 
ts.replace(cleanup_1, inplace=True)
cleanup_2={"Embarked":{"Q":2, "C": 1,"S":0}} 
ts.replace(cleanup_2, inplace=True)
display(ts)
```

```python
x = ts.drop(["Survived", "Name", "Cabin", "Ticket","Embarked"], axis=1)
y = ts["Survived"]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)
t = tree.DecisionTreeClassifier()  
t.fit(x_train, y_train)
y_pred=t.predict(x_test)
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
dot_data = tree.export_graphviz(t,out_file= None)
graph = graphviz.Source(dot_data)
graph.render("data")
graph
```

```python
x = ts.drop(["Survived", "Name", "Cabin", "Ticket","SibSp","Parch","PassengerId"], axis=1)
y = ts["Survived"]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)
t = tree.DecisionTreeClassifier()  
t.fit(x_train, y_train)
y_pred=t.predict(x_test)
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

we can see that by dropping more unnecessary features like the passengerid and adding another useful feature like the embarking place the accuracy became highter

```python
a = ts.drop(["Survived", "Name", "Cabin", "Ticket","SibSp","Parch","PassengerId"], axis=1)
b = ts["Survived"]
a_train, a_test, b_train, b_test= train_test_split(a, b, test_size=0.3)
for i in range(10,50):
    n = 0
    j = 0
    clf=RandomForestClassifier(n_estimators = i)  #Creating a random forest with 100 decision trees
    clf.fit(a_train, b_train)  #Training our model
    b_pred=clf.predict(a_test)  #testing our model
    m = metrics.accuracy_score(b_test, b_pred)
    if (i == 10):
        n = m
    if (m>n):
        n = m
        j = i
print("Accuracy:",n)
print("the number of estimators:",j)
```
