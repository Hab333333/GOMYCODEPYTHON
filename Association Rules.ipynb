```python
import pandas as pd
import numpy as np
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import association_rules 
df = pd.read_csv("C:/Users/DELL/Downloads/Market_Basket_Optimisation.csv",header=None)
display(df)
df_= df.to_numpy().tolist()
df_2 = []
for i in df_:
    c = []
    m = [str(x) for x in i]
    c = [j for j in m if j != 'nan']
    df_2.append(c)
print(df_2)
```

```python
te=TransactionEncoder()
te_ary=te.fit(df_2).transform(df_2)
df_3=pd.DataFrame(te_ary, columns=te.columns_)
display(df_3)
frequent_itemsets=apriori(df_3, min_support=0.05, use_colnames=True)
display(frequent_itemsets)
display(association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2))
display(association_rules(frequent_itemsets,metric="lift",min_threshold=1.25))
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
t=TransactionEncoder()
t_ary=t.fit(dataset).transform(dataset)
df4=pd.DataFrame(t_ary, columns=t.columns_)
display(df4)
frequent_itemset=apriori(df4, min_support=0.4, use_colnames=True)
display(frequent_itemset)
display(association_rules(frequent_itemset,metric="confidence",min_threshold=0.7))
display(association_rules(frequent_itemset,metric="lift",min_threshold=1.25))
```
