```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/DELL/Downloads/CC GENERAL.csv")
df.info()
```

```python
df['MINIMUM_PAYMENTS'].fillna(int(df['MINIMUM_PAYMENTS'].mean()), inplace=True)
df['CREDIT_LIMIT'].fillna(int(df['CREDIT_LIMIT'].mean()), inplace=True)
df.info()
display(df)
```

```python
df=df.drop("CUST_ID", axis=1)
display(df)
```

```python
from sklearn.cluster import AgglomerativeClustering 
model=AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)
agglomerative=pd.DataFrame(clust_labels)
print(agglomerative.value_counts())
agglomerative
```

```python
corr = df.corr()
c = sns.diverging_palette(240, 10, n=9)
plt.figure(figsize = (24,18))
sns.heatmap(corr,cmap=c,square=True,annot = True,annot_kws = { 'fontsize' : 15 },linewidths=1, linecolor='white')
sns. set(font_scale=1.4)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df ['CASH_ADVANCE'] , df ["PURCHASES"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("CASH_ADVANCE")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(25,20))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df, method="complete"))
```

```python
from sklearn.cluster import KMeans
k=KMeans(n_clusters=5, random_state=0) 
k.fit(df)
```

```python
print(k.labels_)
label=pd.DataFrame(k.labels_)
display(label)
```

```python
k.predict(df)
print(k.cluster_centers_)
```

```python
df2 = pd.concat([df, label], axis=1, join='inner')
df2.rename(columns = { 0 :'label'}, inplace = True)
display(df2)
```

```python
plt.scatter(df2["CASH_ADVANCE"][df2.label == 0], 
           df2["PURCHASES"][df2.label == 0],s=80,c='magenta')
plt.scatter(df2["CASH_ADVANCE"][df2.label == 1], 
           df2["PURCHASES"][df2.label == 1],s=80,c='yellow')
plt.scatter(df2["CASH_ADVANCE"][df2.label == 2], 
           df2["PURCHASES"][df2.label == 2],s=80,c='burlywood')
plt.scatter(df2["CASH_ADVANCE"][df2.label == 3], 
           df2["PURCHASES"][df2.label == 3],s=80,c='cyan')
plt.scatter(df2["CASH_ADVANCE"][df2.label == 4], 
           df2["PURCHASES"][df2.label == 4],s=80,c='burlywood')
plt.scatter(k.cluster_centers_[:, 0], k.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('CASH_ADVANCE')
plt.ylabel('PURCHASES')
plt.legend()
plt.show()
```
